# Load in the data

```python
# Price changes 
PriceChange = pd.read_pickle(OutputFolder + 'PriceChange_Sample.p')
# gvkey 
DS_FirmInfo     =   pd.read_pickle("Micro_data/temp/SDC_CompustatQ_FirmInfo.p")
```

```python
# Merge the data 
DS = PriceChange.merge(DS_FirmInfo[['IssueID', 'gvkey']], on='IssueID', how='left', validate='m:1', indicator=True)
```

```python
VarList_PriceChange = ['PC_Total_Event', 'PC_FirstEvent', 'PC_Runup'] + \
                      ['_'.join([xx, yy]) for xx in ['PC_Total_Event', 'PC_FirstEvent'] for yy in ['FF5_PreEvent_F', 'FF5_PostEvent_I']]
VarList_ID  = ['IssueID', 'gvkey']
VarList_Date = [xx+yy for xx in ['FirstEventDate', 'FilingDate', 'LaunchDate', 'IssueDate'] for yy in ['', '_Quarter']]
VarList_Issuance = ['s_Share', 'ShelfIssueFlag']
```

```python
VarList_All = VarList_ID + VarList_Issuance + VarList_Date + VarList_PriceChange
```

```python
Temp = DS.loc[:, VarList_All]
```

```python
obj_cols = Temp.select_dtypes(include=['object']).columns
for col in obj_cols:
    Temp[col] = Temp[col].astype(str)
```

```python
Temp.to_stata('Micro_data/results/DataForPablo/Sample_SEO.dta', write_index=False)
```
